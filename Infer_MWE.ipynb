# ADS Full Inference MWE

Dieses Notebook ist das **MWE** fuer das AND-Package selbst.

Es zeigt nur den normalen Standardpfad:
- Rohdaten rein
- `disambiguate_sources(...)`
- Ergebnis und `summary.json` anschauen

Es nutzt:
- die **rohen ADS-Dateien**
- das **mitgelieferte Fixed Model Baseline**
- `runtime="auto"`, also GPU wenn verfuegbar, sonst CPU
- die **interne** kompakte Default-Progress-UX des AND-Packages


```python
from __future__ import annotations

import json
import sys
from pathlib import Path

REPO_ROOT = Path('/home/ubuntu/Author_Name_Disambiguation')
if str(REPO_ROOT / 'src') not in sys.path:
    sys.path.insert(0, str(REPO_ROOT / 'src'))

from author_name_disambiguation import disambiguate_sources

PUBLICATIONS_PATH = REPO_ROOT / 'data/raw/ads/ads_full_run_20260305/publ_final.parquet'
REFERENCES_PATH = REPO_ROOT / 'data/raw/ads/ads_full_run_20260305/refs_final.parquet'
OUTPUT_DIR = REPO_ROOT / 'artifacts/exports/notebook_infer_mwe'
FORCE = True

PUBLICATIONS_PATH, REFERENCES_PATH, OUTPUT_DIR
```

```python
result = disambiguate_sources(
    publications_path=PUBLICATIONS_PATH,
    references_path=REFERENCES_PATH,
    output_dir=OUTPUT_DIR,
    runtime='auto',
    progress=True,
    progress_style='compact',
    force=FORCE,
)
```

```python
summary = json.loads(result.summary_path.read_text(encoding='utf-8'))
summary
```
