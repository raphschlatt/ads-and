Einfaches Reporting für einen bestehenden Lauf.

```python
from pathlib import Path
import sys

RUN_STAGE = "smoke"  # smoke|mini|mid|full
PATHS_CONFIG = "configs/paths.local.yaml"  # alternativ: configs/paths.colab.yaml

def _find_root(start: Path) -> Path:
    for c in [start, *start.parents]:
        if (c / "src").exists() and (c / "configs").exists():
            return c.resolve()
    return start.resolve()

ROOT = _find_root(Path.cwd())
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
print("ROOT:", ROOT)

RUN_ID = "replace_with_run_id"
RUN_STAGE = "full"
import pandas as pd
```

```python
from src.common.config import load_notebook_context, build_run_dirs

CTX = load_notebook_context(run_stage=RUN_STAGE, project_root=ROOT, paths_config=PATHS_CONFIG)
RUN_DIRS = build_run_dirs(CTX["DATA"], CTX["ART"], RUN_ID)
clusters = pd.read_parquet(RUN_DIRS["clusters"] / f"ads_clusters_{RUN_STAGE}.parquet")
print("rows", len(clusters), "unique_uids", clusters["author_uid"].nunique())
clusters.head()
```

```python
sizes = clusters.groupby("author_uid").size().rename("cluster_size")
print(sizes.describe())
sizes.sort_values(ascending=False).head(20)
```
