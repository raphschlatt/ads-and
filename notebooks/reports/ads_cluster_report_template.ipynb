# ADS Cluster Report Template
Lädt Cluster-Outputs eines `run_id` und zeigt Basisdiagnostik.

```python
from pathlib import Path
import pandas as pd
import sys

RUN_ID = "replace_with_run_id"
RUN_STAGE = "full"
def _find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / "src").exists() and (candidate / "configs").exists():
            return candidate.resolve()
    return start.resolve()

PROJECT_ROOT = _find_project_root(Path.cwd())
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
```

```python
clusters = pd.read_parquet(PROJECT_ROOT / "artifacts/clusters" / RUN_ID / f"ads_clusters_{RUN_STAGE}.parquet")
print("rows", len(clusters), "unique_uids", clusters["author_uid"].nunique())
clusters.head()
```

```python
sizes = clusters.groupby("author_uid").size().rename("cluster_size")
print(sizes.describe())
sizes.sort_values(ascending=False).head(20)
```
